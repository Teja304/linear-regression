```python
#4 Convert the columns uploads, Subscriptions and Views to numeric
df_yt["Uploads"]=df_yt["Uploads"].str.replace(",","")
df_yt["Uploads"]=pd.to_numeric(df_yt["Uploads"])
```

```python
#14 Converting to numeric columns in Views
df_yt["Views"]=df_yt["Views"].str.replace(",","")

#15 Check the rows where views have missing values represented by "--"
df_yt[df_yt["Views"]=="--"]["Views"]

#16 Replace the values by 0
l1=[]
for i in df_yt["Views"]:
    if i=="--":
        i=0
    l1.append(i)

df_yt["Views"]=l1
df_yt["Views"]=df_yt["Views"].str.replace(",","")
df_yt["Views"]=pd.to_numeric(df_yt["Views"])
```

```python
df_yt["Views"]
```

```python
df_yt["Subscriptions"]=df_yt["Subscriptions"].str.extract("(\d+)")
df_yt["Subscriptions"]=pd.to_numeric(df_yt["Subscriptions"])
```

```python
#17 Print total channel subscriptions gradewise
df_group=df_yt.groupby("Grade")["Subscriptions"].sum()
df_group
```

```python
sns.barplot(x=df_group.index,y=df_group)
```

```python
#18 Print a pie plot showing the total number of uploads grade-wise
plt.pie(df_yt.groupby("Grade")["Subscriptions"].sum(),labels=df_yt.groupby("Grade")["Subscriptions"].sum().index)
```
